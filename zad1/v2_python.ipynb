```python
import numpy as np
import random
import matplotlib.pyplot as plt
```

```python
L = 100
network = np.round(np.random.rand(L, L), 0)
i = 0
yesses = []
while (0 < np.sum(network) < L**2) and i<100_000:
    i += 1
    temp = np.zeros((L, L))
    for x in range(L):
        for y in range(L):
            decision = random.randint(1, 4)
            if decision == 1:
                choice = network[L-1 if x-1 < 0 else x-1][y]
            elif decision == 2:
                choice = network[0 if x+1 > L-1 else x+1][y]
            elif decision == 3:
                choice = network[x][L-1 if y-1 < 0 else y-1]
            else:
                choice = network[x][0 if y+1 > L-1 else y+1]
            network[x][y] = choice
    
    yesses.append(np.sum(network))
print(i)
plt.plot(yesses)
plt.show()
```
