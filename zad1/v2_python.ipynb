```python
import numpy as np
import random
import matplotlib.pyplot as plt
```

```python
L = 100
network = np.round(np.random.rand(L, L), 0)
i = 0
yesses = []
while (0 < np.sum(network) < L**2) and i < 20_000:
    i += 1
    temp = np.zeros((L, L))
    for x in range(L):
        for y in range(L):
            decision = random.randint(1, 4)
            if decision == 1:
                temp[x][y] = network[L-1 if x-1 < 0 else x-1][y]
            elif decision == 2:
                temp[x][y] = network[0 if x+1 > L-1 else x+1][y]
            elif decision == 3:
                temp[x][y] = network[x][L-1 if y-1 < 0 else y-1]
            else:
                temp[x][y] = network[x][0 if y+1 > L-1 else y+1]
            
    
    yesses.append(np.sum(temp))
    if i>5:
        if yesses[i-2] == yesses[i-3] == yesses[i-4] == L/2:
            break
    network = temp
print(i)
print(network[1:5, 1:5], "\t...")
print("\t .")
print("\t .")
print("\t .")
plt.plot(yesses)
plt.show()
```
